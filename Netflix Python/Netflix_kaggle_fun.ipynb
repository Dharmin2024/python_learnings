```python
import ast

# Convert string to list if necessary
data['genres'] = data['genres'].apply(ast.literal_eval)

# Extract all unique genres
unique_genres = set(genre for genres_list in data['genres'] for genre in genres_list)
print(unique_genres)
```

```python
from collections import Counter

# Flatten the list of genres and count occurrences
genre_counts = Counter(genre for genres_list in data['genres'] for genre in genres_list)

# Convert to a DataFrame for better readability (optional)
genre_counts_df = pd.DataFrame(genre_counts.items(), columns=['Genre', 'Count']).sort_values(by='Count', ascending=False)
print(genre_counts_df)
```

```python
# Step 1: Expand genres into multiple rows
expanded_data = data.explode('genres')

# Step 2: Group by genres and sort by IMDb score
top_movies_by_genre = (
    expanded_data.sort_values(by='imdb_score', ascending=False)
    .groupby('genres')
    .head(5)
)

# Step 3: Display the results
print(top_movies_by_genre[['genres', 'title', 'imdb_score']].sort_values(['genres', 'imdb_score'], ascending=[True, False]))
```
